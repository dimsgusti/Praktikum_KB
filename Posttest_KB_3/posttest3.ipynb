```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_all_phones.csv")
df
```

```python
data = df.groupby("brand").mean()
data
```

## Grafik Batang

```python
x = data.index
y = data.battery

plt.bar(x,y)
plt.title('Rata-rata baterai setiap brand')
plt.xlabel('Brand')
plt.ylabel('Rata-rata baterai')
plt.xticks(rotation=45)
plt.show()
```

Kesimpulan, Brand Apple paling sedikit memberikan produknya dengan baterai yang besar tidak seperti brand Realme dan Vivo yang memberikan kapasitas baterai hampir 5000mAh

## Histogram

```python
plt.hist(df['price(USD)'], bins=20)
plt.title('Histogram Harga Handphone')
plt.xlabel('Harga HP')
plt.ylabel('Frekuensi Kemunculan')
plt.xticks(rotation=90)
plt.show()
```

Dari semua brand HP yang beredar harga HP paling banyak berkisar di harga 0 - 500 USD dan sangat sedikit hp berharga lebih dari 500 USD

## Pie Chart

```python
jumlah = df.brand.value_counts()

brands = jumlah.index
hitung = jumlah.values

plt.figure(figsize=(10,8))
plt.pie(hitung, labels=brands, autopct="%1.1f%%", startangle=140)
plt.title('Jumlah data')
plt.show()
```

Brand Apple paling sedikit mengeluarkan produk dibandingkan dengan brand Xiaomi yang terlihat pada presentase mendapatkan nilai terbesar yaitu 17.5%

## Heatmap

```python
cor_matrix = data.corr()

plt.figure(figsize=(20, 16))
sns.heatmap(cor_matrix, annot=True, center=0)
plt.show()
```

Terdapat korelasi positif terhadap baterai dan kapasitas penyimpanan, semakin besar baterainya semakin besar juga kapasitas penyimpanan pada handphonenya.
