## Membagi training set dan testing set 70:30

```python
x = df.iloc[:, : -1]
y = df.iloc[:, -1]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)

print('x_train', x_train.shape)
print('y_train', y_train.shape)
print('x_test', x_test.shape)
print('y_test', y_test)
```

## Normalisasi data menggunakan Min Max scaler

```python
min_max = MinMaxScaler()

# Copy
copy = df
dataCopy_normalization = copy[['price(USD)']]

normalization = min_max.fit_transform(dataCopy_normalization[['price(USD)']])

df_normalization = pd.DataFrame(normalization)
df_normalization.columns = ['price(USD)_normalization']

dataCopy_normalization = dataCopy_normalization.join(df_normalization['price(USD)_normalization']).reset_index()
dataCopy_normalization = dataCopy_normalization.drop('index', axis=1)

print()
print('Setelah normalization: \n')
dataCopy_normalization.head(10)
```

## Standarisasi

```python
copy2 = df.copy()

attribute_to_standardize = 'price(USD)'

scaler = StandardScaler()

copy2[attribute_to_standardize] = scaler.fit_transform(copy2[[attribute_to_standardize]])

print('Attribute sebelum standarisasi: \n')
print(df[[attribute_to_standardize]].head(10))
print('Attribute sesudah standarisasi: \n')
print(copy2[[attribute_to_standardize]].head(10))
```

## Data cleaning

```python
# Menampilkan nilai null
df.isna().sum()
```

```python
# Menghapus nilai null
df.dropna(inplace=True)
df.isna().sum()
```

```python
# Mengubah bilangan bulat menjadi median/modus
df2 = pd.read_csv('cleaned_all_phones.csv')
strategi = SimpleImputer(strategy='median')

df2['price(USD)'] = strategi.fit_transform(df2[['price(USD)']])

df2.isna().sum()
```

```python
# Mengubah bilangan desimal menjadi mean
column_to_impute = 'weight(g)'

imputer = SimpleImputer(strategy='mean')

df[column_to_impute] = imputer.fit_transform(df[[column_to_impute]])

missing_values = df.isna().sum()

print(missing_values)
```

```python
# Menghubah tulisan menjadi modus
# Misalnya, kita mengganti nilai null dengan median, mean, atau modus sesuai dengan jenis atributnya
df.fillna({'categorical_column': df['weight(g)'].mode()[0]}, inplace=True)
df.isna().sum()
```

## Data cleaning pada nilai duplikat

```python
duplicate_data = df[df.duplicated(keep=False)]
duplicate_data
```

```python
jumlah_duplikat = 10
duplicate = pd.concat([df] * jumlah_duplikat, ignore_index=True)
```
